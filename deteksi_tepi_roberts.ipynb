```python
# Import library

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image

img = cv2.imread('grayscale-lenna.png',0)
```

```python
# Convert from BGR to RGB

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
# Show the image

plt.imshow(img)
```

```python
# histogram equalization

#img = cv2.equalizeHist(img)
```

```python
# Setting Gx (for vertikal edge) and Gy (for horizontal edge)

Gx = np.array([[1, 0], [0, -1]])
Gy = np.array([[0, 1], [-1, 0]])
```

```python
Gx
```

```python
Gy
```

```python
# Running convolution

Gx_new = cv2.filter2D(img, -1, Gx)
Gy_new = cv2.filter2D(img, -1, Gy)
```

```python
# the sum of the squares of the differences between diagonally adjacent pixels

Gr_new = abs(Gx_new) + abs(Gy_new)
```

```python
plt.imshow(Gr_new)
```

```python
# Kernel from Sobel

Gxs = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
Gys = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
```

```python
# Running convolution

Gxs_new = cv2.filter2D(img, -1, Gxs)
Gys_new = cv2.filter2D(img, -1, Gys)
```

```python
Gs_new = abs(Gxs_new) + abs(Gys_new)
```

```python
plt.imshow(Gs_new)
```

```python
# Kernel from Prewitt

Gxp = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
Gyp = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
```

```python
# Running convolution

Gxp_new = cv2.filter2D(img, -1, Gxp)
Gyp_new = cv2.filter2D(img, -1, Gyp)
```

```python
Gp_new = abs(Gxp_new) + abs(Gyp_new)
```

```python
plt.imshow(Gp_new)
```

```python
# Kernel from Canny

img_can = cv2.Canny(img, 200, 300)
```

```python
plt.imshow(img_can)
```
